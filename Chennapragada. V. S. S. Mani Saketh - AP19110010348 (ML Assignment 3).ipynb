1. Take two 3D coordinates from the user. Find out the (i) Euclidean distance, (ii) Manhattan distance and (iii) Minkowski distance between these points.

```python
#1
import math
from math import sqrt
l1 = []
l2 = []
print("Enter Point 1 elements")
for i in range(3):
    num = int(input())
    l1.append(num)
print("Enter Point 2 elements")
for i in range(3):
    num = int(input())
    l2.append(num)
#euclidean
#manhattan
#minkowski
f_part = 0
s_part = 0
t_part = 0
print("Enter p Value")
p = int(input())
for i in range(3):
    f_part += pow(l1[i] - l2[i],2)
    s_part += abs(l1[i] - l2[i])
    t_part += pow(abs(l1[i] - l2[i]),p)
euclidean = sqrt(f_part)
manhattan = s_part
minkowski = pow(t_part,1/p)
print("Euclidean Distance is",euclidean)
print("Manhattan Distance is",manhattan)
print("Minkowski Disance is",minkowski)
```
